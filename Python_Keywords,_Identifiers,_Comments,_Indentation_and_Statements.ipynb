## 4. Why is proper indentation important in Python?

Python relies on indentation to determine which lines of code belong together. Consistent indentation levels tell the Python interpreter the program structure, differentiating between code within loops, condtional statements, functions, and classes.    
Proper indentation also makeit easier to understan. By clearly indicating which statements belongs to which block, indented code creates a hierachy that reflects the program's logic. This is especillary belfically for complex code and nested loops and conditional statement.
